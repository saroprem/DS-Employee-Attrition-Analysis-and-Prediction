# 1. YEARS WISE RICE PRODUCTION (TOP 3 STATE )

```python
query1 = """
SELECT Year, State_Name,
       SUM(Rice_Production) AS Total_Rice
FROM crop_data
GROUP BY Year, State_Name
ORDER BY Year ASC, Total_Rice DESC;
"""

pd.read_sql_query(query1, conn)
```

# 2. TOP 5 DISTRICTS WHEAT YIELD 5 YEARS INCREASE

```python
query2 = """
SELECT Dist_Name,
       AVG(Wheat_Yield) AS Avg_Wheat_Yield
FROM crop_data
WHERE Year >= (SELECT MAX(Year) - 5 FROM crop_data)
GROUP BY Dist_Name
ORDER BY Avg_Wheat_Yield DESC
LIMIT 5;
"""

pd.read_sql_query(query2, conn)
```

# 3. STATES WITH HIGHEST GROWTH IN OILSEED PRODUCTION (LAST 5 YEARS)

```python
query3 = """
SELECT State_Name,
       (MAX(Oilseed_Production) - MIN(Oilseed_Production)) AS Oilseed_Growth
FROM crop_data
WHERE Year >= (SELECT MAX(Year) - 5 FROM crop_data)
GROUP BY State_Name
ORDER BY Oilseed_Growth DESC
LIMIT 5;
"""

pd.read_sql_query(query3, conn)
```

# 4. AREA WISH PRODUCTION CORRELATION (RICE, WHEAT,MAIZE)

```python
query4 = """
SELECT 
    Rice_Area, Rice_Production,
    Wheat_Area, Wheat_Production,
    Maize_Area, Maize_Production
FROM crop_data;
"""

df_corr = pd.read_sql_query(query4, conn)
df_corr.head()
```

# EARLY PRODUCTION GROWTH OF COTTON (TOP STATES )

```python
query5 = """
SELECT Year, State_Name,
       SUM(Cotton_Production) AS Total_Cotton
FROM crop_data
GROUP BY Year, State_Name
ORDER BY Year ASC, Total_Cotton DESC;
"""

pd.read_sql_query(query5, conn)
```

# 6. DISTRICTS WITH HIGHEST GROUNDNUT PRODUCTION (YEAR=2020)

```python
query6 = """
SELECT Dist_Name, Groundnut_Production
FROM crop_data
WHERE Year = 2020
ORDER BY Groundnut_Production DESC
LIMIT 10;
"""

pd.read_sql_query(query6, conn)
```

# 7. ANNUAL AVERAGE MAIZE YIELD (ACROSS ALL STATES)

```python
query7 = """
SELECT Year,
       AVG(Maize_Yield) AS Avg_Maize_Yield
FROM crop_data
GROUP BY Year
ORDER BY Year ASC;
"""

pd.read_sql_query(query7, conn)
```

# 8. TOTAL OILSEED PRODUCTION BY STATE

```python
query8 = """
SELECT State_Name,
       SUM(Oilseed_Production) AS Total_Oilseed_Production
FROM crop_data
GROUP BY State_Name
ORDER BY Total_Oilseed_Production DESC;
"""

pd.read_sql_query(query8, conn)
```

# 9. TOP DISTRICTS WITH HIGHEST RICE YIELD

```python
query9 = """
SELECT Dist_Name, Rice_Yield
FROM crop_data
ORDER BY Rice_Yield DESC
LIMIT 10;
"""

pd.read_sql_query(query9, conn)
```

# 10. COMPARE WHEAT VS RICE PRODUCTION FOR TOP STATES (10 YEARS)

```python
query10 = """
SELECT Year,
       State_Name,
       SUM(Rice_Production) AS Rice_Prod,
       SUM(Wheat_Production) AS Wheat_Prod
FROM crop_data
GROUP BY Year, State_Name
ORDER BY Year ASC, Rice_Prod DESC, Wheat_Prod DESC;
"""

pd.read_sql_query(query10, conn)
```
